```python
import cv2 ##Open CV Library


image = cv2.imread("Images\people2.jpg")
if len(image.shape) == 3:
    image_to_show = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
else:
    image_to_show = image
```

```python
image_to_show = cv2.resize(image_to_show,(800,600))
```

```python
image_to_show.shape
```

```python

image_gray = cv2.cvtColor(image_to_show,cv2.COLOR_BGR2GRAY)

```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(image_gray,cmap='gray')
plt.axis('off')
plt.show()
```

```python
image_gray.shape
```

```python
##Detecting Face

face_detector = cv2.CascadeClassifier('XML Files\haarcascade_frontalface_default.xml')
```

```python
detection = face_detector.detectMultiScale(image_gray,minSize=(20,20),maxSize=(100,100))
detection
```

```python
len(detection)
```

```python
for (x,y,w,h) in detection:
    cv2.rectangle(image_to_show, (x,y),(x+w,y+h),(0,255,155),5)
plt.imshow(image_to_show)
plt.axis('off')
plt.show()
```
